```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert_left(self, node):
        if self.data:
            if self.left is None:
                self.left = node
            else:
                self.left.insert_left(node)
        else:
            self.data = node.data

    def insert_right(self, node):
        if self.data:
            if self.right is None:
                self.right = node
            else:
                self.right.insert_right(node)
        else:
            self.data = node.data

    def __str__(self):
        return str(self.data)
```

```python
def curlybracketcounter(code, i):
    j = i
    stack = []
    while j>(i-1):
        if('{' in code[j]):
            stack.append('{')
            #print(stack)
        if('}' in code[j] and '{' in stack):
            stack.remove('{')
            #print(stack)
        if(len(stack) == 0):
            k = j
            #print(k)
            break
        j+=1
    return(k)
```

```python
def ASTgenerator(file):
    file = open(file , 'r')
    imp_lines = []
    for line in file:
        if('//' in line):
            line = line[0:line.index("//")]
            #print(line)
        if('for' in line or 'while' in line):
            line = line.replace(";",":")
        if(not line.startswith('//') and line!="\n" and not line.startswith('#') and not line.startswith('using')):
            imp_lines.append(line.lower().strip())

    code = ""
    for i in range(len(imp_lines)):
        code += imp_lines[i]

    code = code.translate({ord("}"):"};",ord("("):" ( ",ord("{"):" { ;"})
    lines_code = code.strip().split(';')
    #lines_code.remove('')

    Lines_code = lines_code

    AST = []
    i = 0
    while i <= (len(Lines_code)-1):
    #print(Lines_code[i])
        if('{'in Lines_code[i] and not('while' in Lines_code[i]) and not('for' in Lines_code[i])):
            ans = func(Lines_code,i)
            f = ans[0]
            AST.append(f)
            #print(AST)
            i = ans[1]
            #print(i)
            continue
        if('cout' in Lines_code[i] or 'cin' in Lines_code[i] or 'break' in Lines_code[i]):
            AST.append(Node('I'))
            #print(AST)
            i+=1
            #print(i)
            continue
        if('=' in Lines_code[i]):
            a = Lines_code[i].find('=')
            if((a-1)>1):
                AST.append(Node('D'))
                #print(AST)
            else:
                AST.append(Node('A'))
                #print(AST)
            i+=1
            #print(i)
            continue
        if('(' in Lines_code[i] and ')' in Lines_code[i] and not('{' in Lines_code[i])):
            AST.append(Node('I'))
            #print(AST)
            i+=1
            #print(i)
            continue
        if('}'in Lines_code[i]):
            i+=1
            continue
        if(not('{'in Lines_code[i]) or not('}'in Lines_code[i]) or not('for'in Lines_code[i]) or not('while'in Lines_code[i])
        or not('else'in Lines_code[i]) or not('return'in Lines_code[i]) or not('('in Lines_code[i])or not(')'in Lines_code[i])
        or not('='in Lines_code[i]) or not('if'in Lines_code[i]) or not('if'in Lines_code[i]) or not('elseif'in Lines_code[i])
        or not('switch'in Lines_code[i]) or not('cout'in Lines_code[i])or not('cin'in Lines_code[i]) or not('break'in Lines_code[i])):
            #print(i)
            #print(Lines_code[i])
            AST.append(Node('D'))
            i+=1
            continue
        if i < len(Lines_code):
            i+=1
            continue
    #print(AST)
    return(AST)
```

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.data = key

class BinaryTree:
    def __init__(self):
        self.root = None

    def addNode(self, node, value):
        if(node==None):
            self.root = Node(value)
        else:
            if(value<node.data):
                if(node.left==None):
                    node.left = Node(value)
                else:
                    self.addNode(node.left, value)
            else:
                if(node.right==None):
                    node.right = Node(value)
                else:
                    self.addNode(node.right, value)

def nodes_of_tree(root,nodes_list=[]):
    if root:
        nodes_of_tree(root.left,nodes_list)
        nodes_list.append(root)
        nodes_of_tree(root.right,nodes_list)
    return nodes_list

def preOrder(root,list_preorder=[]):
    if root:
        list_preorder.append(root.data)
        #print(root.val)
        preOrder(root.left,list_preorder)
        preOrder(root.right,list_preorder)
    return list_preorder

def inOrder(root,list_inorder=[]):
    if root:
        inOrder(root.left,list_inorder)
        list_inorder.append(root.data)
        #print(root.val)
        inOrder(root.right,list_inorder)
    return list_inorder

def postOrder(root,list_postorder=[]):
    if root:
        list_postorder.append(root.data)
        postOrder(root.left,list_postorder)
        postOrder(root.right,list_postorder)
    return list_postorder
```
